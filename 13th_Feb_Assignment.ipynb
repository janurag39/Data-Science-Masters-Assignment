### Q1. Explain why we have to use the Exception class while creating a Custom Exception.
- Note: Here Exception class refers to the base class for all the exceptions.

Ans. It is important to use Exception class while creating custom exceptions/User defined exceptions, because we need to make sure that our custom exception inherits all the properties and methods from the base Exception class that consist of all python errors.
- Ex: raise keyword can only be used when custom exception inherits the base Exception class method and properties.

### Q6. List down some best practices for exception handling in python. 

Ans.
1. Always use specific exceptions/errors while creating except block instead of Exception superclass.
2. Print always a valid Exception message that is understandable by others as well.
3. Always log the error messages for future refereces.
4. Always avoid writing multiple exception handlings when they aren't even relateable to the code
5. Always prepare a proper documentation.
