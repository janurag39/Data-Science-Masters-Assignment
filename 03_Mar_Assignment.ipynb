### Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Ans.
1. **relplot()**: The function relplot() is named that way because it is designed to visualize many different statistical relationships. While scatter plots are often effective, relationships where one variable represents a measure of time are better represented by a line.
2. **displot()**: Statistical analyses require knowledge about the distribution of variables in your dataset. The seaborn function displot() supports several approaches to visualizing distributions. These include classic techniques like histograms and computationally-intensive approaches like kernel density estimation.
3. **catplot()**: Several specialized plot types in seaborn are oriented towards visualizing categorical data. They can be accessed through catplot(). These plots offer different levels of granularity. At the finest level, you may wish to see every observation by drawing a “swarm” plot: a scatter plot that adjusts the positions of the points along the categorical axis so that they don’t overlap.
4. **jointplot()**: Some seaborn functions combine multiple kinds of plots to quickly give informative summaries of a dataset. One, jointplot(), focuses on a single relationship. It plots the joint distribution between two variables along with each variable’s marginal distribution.
5. **pairplot()**: It shows joint and marginal distributions for all pairwise relationships and for each variable, respectively.
6. **lineplot()**: It is used to plot the line graph.

### Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions. 

Note:  timepoint, signal, event, and region are columns in the fmri dataset.

```python
import seaborn as sns

fmri = sns.load_dataset("fmri")
```

```python
fmri.columns
```

```python
sns.lineplot(x = fmri["timepoint"], y = fmri["signal"], hue=fmri["region"], style=fmri["event"])
```

### Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 'pclass', y = 'age' and y = 'fare'.

Note:  pclass, age, and fare are columns in the titanic dataset

```python
import seaborn as sns
import matplotlib.pyplot as plt

t = sns.load_dataset("titanic")
```

```python
t.columns
```

```python
fig, axes = plt.subplots(1,2)
sns.boxplot(x="pclass", y="age", data=t, ax=axes[0])
sns.boxplot(x="pclass", y="fare", data=t, ax=axes[1])
```

### Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset. 

```python
import seaborn as sns

dm = sns.load_dataset("diamonds")
```

### Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
import seaborn as sns
import matplotlib.pyplot as plt

fg = sns.load_dataset("flights")
```

```python
fg.head()
```

```python
fg = fg.pivot(index="month", columns="year", values="passengers")
```

```python
vis = sns.heatmap(fg)
plt.title("Heatmap Flight Data")
plt.show()
```
