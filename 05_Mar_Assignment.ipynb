### Q1. How can you create a Bokeh plot using Python code?

```python
import bokeh.io
import bokeh.plotting
from bokeh.plotting import figure, show, output_file
from bokeh.sampledata.iris import flowers

bokeh.io.output_notebook() # to print graph on notebook
```

```python
output_file("Ques1.html")
fig1 = figure(title = "test flower")
fig1.xaxis.axis_label = "Petal Length"
fig1.yaxis.axis_label = "Petal Width"
fig1.scatter(flowers["petal_length"], flowers["petal_width"])
show(fig1)
```

### Q2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.

Ans. With the help of Glyphs we can Display a variety of visual shapes whose attributes can be associated with data of given columns in the dataframe. 

Basic Glyphs can be used in the plot with the name instead of plot-name: for ex: if we want to plot diamond we use figure.diamond instead of figure.scatter or any plot name as such

```python
import bokeh.io
import bokeh.plotting
from bokeh.plotting import figure, show, output_file

bokeh.io.output_notebook() # to print graph on Notebook
```

```python
x = [1, 2, 3, 4, 5]
y = [10,11,12,14,15]
output_file("Ques2.html")
fig2 = figure(title="Example for Glyphs in Bokeh")
fig2.diamond(x,y, size=30)
show(fig2)
```

### Q3. How can you customize the appearance of a Bokeh plot, including the axes, title, and legend?

```python
from bokeh.core.property.numeric import Size
import bokeh.io
import bokeh.plotting
from bokeh.plotting import figure, show, output_file
from bokeh.sampledata.iris import flowers
from bokeh.palettes import viridis

bokeh.io.output_notebook() # to print graph on Notebook

output_file("Ques2.html")
fig3 = figure(title="Example for Glyphs in Bokeh") # for title
# for axis labels
fig3.xaxis.axis_label = "X Axis" 
fig3.yaxis.axis_label = "Y Axis"
fig3.scatter(flowers["petal_length"], flowers["sepal_length"], legend_label="Circle plot", color=viridis(256))
fig3.legend.location = 'top_left' # legend position
fig3.legend.title = 'Legend Title' # legend title
fig3.legend.title_text_font = 'Arial' # lengend font
fig3.legend.title_text_font_size = '10pt' # legend text font size
show(fig3)
```

### Q4. What is a Bokeh server, and how can you use it to create interactive plots that can be updated in real time?

Ans. The **Bokeh server** is a component of Bokeh that allows us to build interactive web applications that are connected to Python code running on a server.

The most flexible way to create interactive data visualizations with the Bokeh server is to create Bokeh applications and serve them with the bokeh serve command. The Bokeh server then uses the application code to create sessions and documents for all connecting browsers. The Bokeh server executes the application code with every new connection and creates a new Bokeh document, syncing it to the browser. The application code also sets up the callbacks that should run whenever properties, such as widget values, change.
