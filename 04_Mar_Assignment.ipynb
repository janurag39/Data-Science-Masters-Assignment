### Q1. Load the "titanic" dataset using the load_dataset function of seaborn. Use Plotly express to plot a scatter plot for age and fare columns in the titanic dataset.

```python
import seaborn as sns

# Using seaborn to import "titanic" dataset
t = sns.load_dataset("titanic")
t.columns
```

```python
import plotly.graph_objs as go

fig1 = go.Figure()
fig1.add_trace(go.Scatter(x=t.age, y=t.fare, mode="markers"))
fig1.update_xaxes(title_text="Age (in Years)")
fig1.update_yaxes(title_text="Fare (in Euros)")
```

### Q2. Using the tips dataset in the Plotly library, plot a box plot using Plotly express.

```python
import plotly.express as px
import pandas as pd

tips = px.data.tips()
tips.head()
```

```python
fig2 = px.box(data_frame=tips, x="time", y="total_bill", labels={"time": "Time (Dinner/Lunch)", "total_bill": "Total Bill" }, title="Box Graph (Time vs Total Bill)")
fig2.show()
```

### Q3. Using the tips dataset in the Plotly library, Plot a histogram for x= "sex" and y="total_bill" column in the tips dataset. Also, use the "smoker" column with the pattern_shape parameter and the "day" column with the color parameter.

```python
import plotly.express as px

tips = px.data.tips()
tips.head()
```

```python
fig3 = px.histogram(data_frame=tips, x="sex", y="total_bill", pattern_shape="smoker", color="day", labels={"sex": "Sex (Male/Female)", "total_bill": "Total Bill", "smoker": "Smoker", "day": "Day" }, title="Histogram (Sex vs Total Bill)")
fig3.show()
```

### Q4. Using the iris dataset in the Plotly library, Plot a scatter matrix plot, using the "species" column for the color parameter.

Note: Use "sepal_length", "sepal_width", "petal_length", "petal_width" columns only with the 
dimensions parameter.

```python
import plotly.express as px

iris = px.data.iris()
iris.head()
```

```python
fig4 = px.scatter_matrix(iris, color="species", height=900, dimensions=["sepal_length", "sepal_width", "petal_length", "petal_width"])
fig4.show()
```

### Q5. What is Distplot? Using Plotly express, plot a distplot

Ans. **Distplot**: It is used basically for univariant set of observations and visualizes it through a histogram i.e. only one observation and hence we choose one particular column of the dataset.

```python
import numpy as np
import plotly.figure_factory as ff

np.random.seed(1)
x = np.random.randint(1,100,(1000))
```

```python
fig = ff.create_distplot(hist_data=[x], group_labels=["DistPlot"])
fig.show()
```
